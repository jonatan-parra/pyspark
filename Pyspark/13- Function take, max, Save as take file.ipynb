```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
# Punto de entrada para crear los RDD
sc = spark.sparkContext
```

```python
# take -> Toma los n primeros elementos del RDD
rdd = sc.parallelize('La programación es bella'.split(' '))
rdd.take(2)
```

```python
rdd.take(3)
```

```python
# max
rdd1 = sc.parallelize([item/(item + 1) for item in range(10)])
rdd1.max()
```

```python
rdd1.collect()
```

```python
# saveAsTextFile
rdd.collect()
```

```python
rdd.saveAsTextFile('./rdd')
rdd.coalesce(1).saveAsTextFile('./rdd1')
```
