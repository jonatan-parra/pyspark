```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
# Punto de entrada para crear los RDD
sc = spark.sparkContext
```

```python
# Crear RDD
rdd = sc.parallelize([item for item in range(10)])
```

```python
# Crear variable broadcast
# Estas ocupan memoria en todos los ejecutores
uno = 1
br_uno = sc.broadcast(uno)
```

```python
# Usar variable broadcast
rdd1 = rdd.map(lambda x: x + br_uno.value)
```

```python
rdd1.collect()
```

```python
# Liberar variables broadcast de todos los ejecutores
# Si la variable se volviera a utilizar los datos se retrasnmitirian a los
# ejecutores para que se vuelan a utilizar
br_uno.unpersist()
```

```python
# Se vuelve a llamar la variable broadcast
rdd1  = rdd.map(lambda x: x + br_uno.value)
rdd1.collect()
```

```python
# Destrucción de la variable broadcast
# Una vez destruida no se puede volver a utilizar y tendrá que volver a crearse
br_uno.destroy()
```
