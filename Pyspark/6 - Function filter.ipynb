```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
# Punto de entrada para crear los RDD
sc = spark.sparkContext

# Crea un RDD paralelizado
rdd = sc.parallelize([1,2,3,4,5,6,7,8,9])
```

```python
# RDD con números pares
rdd_par = rdd.filter(lambda x: x % 2 == 0)
rdd_par.collect()
```

```python
# RDD con números impares
rdd_impar = rdd.filter(lambda x: x % 2 != 0)
rdd_impar.collect()
```

```python
# RDD de texto - Filtrar que empiecen con K
rdd_texto = sc.parallelize(['jose', 'juaquin', 'juan', 'lucia', 'karla', 'katia'])
rdd_k = rdd_texto.filter(lambda x: x.startswith('k'))
rdd_k.collect()
```

```python
# RDD que empiecen con J y que en su segunda letra tengan una U
rdd_filtro = rdd_texto.filter(lambda x: x.startswith('j') and x.find('u') == 1)
rdd_filtro.collect()
```
