```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
# Punto de entrada para crear los RDD y acumuladores
sc = spark.sparkContext
```

```python
# Crear acumulador
# El parametro es el valor con el cual se desea iniciar el acumulador
acumulador = sc.accumulator(0)
```

```python
# Crear RDD
rdd = sc.parallelize([2,4,6,8,10])
```

```python
# Sumar todos los valores que están en el RDD
rdd.foreach(lambda x: acumulador.add(x))
print(acumulador.value)
```

```python
# Crear RDD
rdd1 = sc.parallelize('Mi nombre es Jose Miguel y me siento genial'.split(' '))
```

```python
# Contar la cantidad de elementos que tiene el RDD usando un acumulador
acumulador1 = sc.accumulator(0)
rdd1.foreach(lambda x: acumulador1.add(1))
print(acumulador1.value)
```
