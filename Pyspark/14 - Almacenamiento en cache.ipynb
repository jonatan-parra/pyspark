```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
# Punto de entrada para crear los RDD
sc = spark.sparkContext
```

```python
# Crear RDD
rdd = sc.parallelize([item for item in range(10)])
```

```python
# Importar la librería para persistir
from pyspark.storagelevel import StorageLevel
```

```python
# Aquí se elige el tipo de persitencia. En este caso en memoria
rdd.persist(StorageLevel.MEMORY_ONLY)
```

```python
# Para cambiar el tipo de persistencia es necesario aplicar unpersist
rdd.unpersist()
```

```python
# Aquí se elige el tipo de persitencia. En este caso en Disco
rdd.persist(StorageLevel.DISK_ONLY)
```

```python
rdd.unpersist()
```

```python
# Cache sinonimo de memory only
rdd.cache()
```
