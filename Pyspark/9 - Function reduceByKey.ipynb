```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
# Punto de entrada para crear los RDD
sc = spark.sparkContext
```

```python
# Crear y mostrar RDD
rdd = sc.parallelize(
    [('casa', 2),
     ('parque', 1),
     ('que', 5),
     ('casa', 1),
     ('escuela', 2),
     ('casa', 1),
     ('que', 1)]
)

rdd.collect()
```

```python
# Reducir por llave
rdd_reducido = rdd.reduceByKey(lambda x,y: x + y)
rdd_reducido.collect()
```
