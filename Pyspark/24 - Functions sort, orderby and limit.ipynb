```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
from pyspark.sql.functions import col
```

```python
# Crear dataframe, selecciona 4 columnas y elimina duplicados usando la columna video_id
# Averiguar como se imitar el keep='first' pyspark no lo tiene

df = (spark.read.parquet('/content/drive/MyDrive/Colab Notebooks/Pyspark/data/datos.parquet')
    .select(col('likes'), col('views'), col('video_id'), col('dislikes'))
    .dropDuplicates(['video_id'])
)

df.show()
```

```python
# Function sort
# Por defecto ordena de menor a mayor
df.sort('likes').show()
```

```python
# Se importa librería para poder ordernar de mayor a menor
from pyspark.sql.functions import desc
df.sort(desc('likes')).show()
```

```python
# función orderBy
df.orderBy(col('views')).show()
```

```python
# Se usar .desc para ordernar de mayor a menor (función .desc fue importada arriba)
df.orderBy(col('views').desc()).show()
```

```python
# Crear dataframe
dataframe = spark.createDataFrame([(1, 'azul', 568), (2, 'rojo', 235), (1, 'azul', 456), (2, 'azul', 783)]).toDF('id', 'color', 'importe')
dataframe.show()
```

```python
# Ordenar de forma decreciente por color, y ascendente por el importe
dataframe.orderBy(col('color').desc(), col('importe')).show()
```

```python
# Function limit
top_10 = df.orderBy(col('views').desc()).limit(10)
top_10.show()
```
