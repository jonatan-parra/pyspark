```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#df = spark.read.parquet('./data/dataPARQUET.parquet')
df = spark.read.parquet('/content/drive/MyDrive/Colab Notebooks/Pyspark/data/datos.parquet')
```

```python
df.printSchema()
```

```python
# Function distinct
df_sin_duplicados = df.distinct()
```

```python
# Realizar conteo para vertificar cuantos datos quedaron
print('El conteo del dataframe original es {}'.format(df.count()))
print('El conteo del dataframe sin duplicados es {}'.format(df_sin_duplicados.count()))
```

```python
# Otra alternativa para imprimir el conteo sin usar .format
print(f'El conteo del dataframe original es {df.count()}')
print(f'El conteo del dataframe sin duplicados es {df_sin_duplicados.count()}')
```
